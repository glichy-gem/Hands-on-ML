# Handling Categorical Data

***1.Manual Conversion***

```python
df1=df.copy(deep=True)
mapping={"Iris-setosa":3, "Iris-versicolor":2, "Iris-virginica":1}
df1["Species"]=df1["Species"].map(mapping)
```

```python
df1.head()
```

```python
df.head()
```

***1.Label Encoder***

```python
df2=df.copy(deep=True)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(df2["Species"])
y
```

```python
df.head()
```

```python
actual=le.inverse_transform(y)
actual
```

***2.One Hot Encoding***

```python
df3=df.copy(deep=True)
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
x3=df3[["Species"]]
ohe.fit_transform(x3).toarray()
```
